# **Managing of missing values**....

```python
# drop the cabin columnfrom the dataframe
df=df.drop(columns='Cabin', axis=1)
```

```python
#replacing the missing values in Age column with mean value
df['Age'].fillna(df['Age'].mean(),inplace=True)
```

```python
# finding the mode value of Embarked Column
print(df['Embarked'].mode())
```

```python
print(df['Embarked'].mode()[0])
```

```python
#replacing the missing values in Embarked column with mode value
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

```python
# check the number of missing values in each column again
df.isnull().sum()
```

# **Converting Categorical columns into numerical values**

```python

df['Sex'].value_counts()
```

```python

df['Embarked'].value_counts()
```

```python
# converting categorical columns
df.replace({'Sex':{'male':0 , 'female':1}, 'Embarked':{'S':0, 'C':1, 'Q':2}} , inplace=True)
```

```python
df.head()
```

# **Saparating Feature and target....**

```python
X = df.drop(columns = ['PassengerId','Name','Ticket','Survived'],axis=1)
Y = df['Survived']
```

```python
print(X)
```

```python
print(Y)
```

# **Spliting the data into test data.....**

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

# **Model Training....**

**Logistic Regression Model:** it is very useful for binary classificatio problem.


Screenshot (151).png

```python
model = LogisticRegression()
```

```python
# training the Logistic Regression model with training data
model.fit(X_train, Y_train)
```

# **Model Evaluation**

**Accuracy Score:**

```python
# accuracy on training dataset
X_train_prediction = model.predict(X_train)
```

```python
print(X_train_prediction)
```

```python
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
print('Accuracy score of training data:' , training_data_accuracy)
```

```python
# accuracy on training dataset
X_test_prediction = model.predict(X_test)
```

```python
print(X_test_prediction)
```

```python
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
print('Accuracy score of testing data:' , test_data_accuracy)
```

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(max_depth=3)

rfc.fit(X_train, Y_train)
```

```python
rfc.score(X_train,Y_train)
```

```python
rfc.score(X_test,Y_test)
```
